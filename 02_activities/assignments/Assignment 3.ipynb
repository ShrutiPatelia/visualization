```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv('/Users/shruti/Desktop/DSI/visualization/02_activities/Building Permits - Pool Enclosures Data.csv')
```

```python
df['ISSUED_DATE'] = pd.to_datetime(df['ISSUED_DATE'], errors='coerce')
```

```python
df['YEAR'] = df['ISSUED_DATE'].dt.year
```

```python
print(df['ISSUED_DATE'].dtype)
print(df['ISSUED_DATE'].head())
```

```python
df = df.dropna(subset=['ISSUED_DATE'])
```

```python
df['Year'] = df['ISSUED_DATE'].dt.year
```

```python
permits_per_year = df['Year'].value_counts().sort_index()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=permits_per_year.index, 
             y=permits_per_year.values,
             marker='o')
plt.title('Annual Number of Pool Enclosure Permits Issued')
plt.xlabel('Year')
plt.ylabel('Number of Permits')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=permits_per_year.index, 
             y=permits_per_year.values,
             color = 'green',
             marker='o')
plt.title('Annual Number of Pool Enclosure Permits Issued')
plt.xlabel('Year')
plt.ylabel('Number of Permits')
plt.tight_layout()
plt.show()
```
